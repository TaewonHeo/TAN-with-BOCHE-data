```python
import pandas as pd
import numpy as np
data = pd.read_csv('/media/D/Backup/Data/train_numeric.csv', nrows=500)
ddata = pd.read_csv('/media/D/Backup/Data/train_date.csv', nrows=500)

col_list = data.columns.tolist()
dcol_list = ddata.columns.tolist()
dcol_list[1:]
ddata
```

```python
ID934 = np.array(ddata.loc[462])[1:] #[1:] ignore 1st row
ID935 = np.array(ddata.loc[461])[1:]

runlen = dict()

for ind, hr in enumerate(ID934):
    if np.isnan(hr):
        pass
    else:
        if hr in runlen:
            runlen[hr].append(dcol_list[ind+1])
        else:
            runlen[hr] = [dcol_list[ind+1]]
runlen
```

```python
L0_dict = dict()
L1_dict = dict()
L2_dict = dict()
L3_dict = dict()

for col in col_list:
    line = col.split('_')[0]
    station = col.split('_')[1]
    if line == "L0":
        if station in L0_dict:
            L0_dict[station] += 1
        else:
            L0_dict[station] = 1
    if line == "L1":
        if station in L1_dict:
            L1_dict[station] += 1
        else:
            L1_dict[station] = 1
    if line == "L2":
        if station in L2_dict:
            L2_dict[station] += 1
        else:
            L2_dict[station] = 1
    if line == "L3":
        if station in L3_dict:
            L3_dict[station] += 1
        else:
            L3_dict[station] = 1
```
